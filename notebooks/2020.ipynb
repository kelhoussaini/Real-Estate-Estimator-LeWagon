```python
def attribute_dependance_column_value(id_mutation, dependences):
    if id_mutation in dependences.id_mutation.values:
        return 1
    else:
        return 0
```

```python
def filter_dependency(data, n_line='all'):
    """
    parameters
    ----------
        data: intitial dataframe 
        
    returns
    -------
        series with type_local filtered by dependency (downcoast to int8)
    """
    
    df_gr= data.copy()
    df_gr = df_gr[df_gr['type_local'] == 'Dépendance']
    if n_line == 'all':
        df=data.apply(lambda x: 1 if x.id_mutation in df_gr.id_mutation.values else 0, axis=1)
        rslt_df = data[(data['type_local'] != 'Dépendance')]
        return rslt_df.astype('int8')
    
    else:
        df=data.head(n_line).apply(lambda x: 1 if x.id_mutation in df_gr.id_mutation.values else 0, axis=1)
        rslt_df = data[(data['type_local'] != 'Dépendance')]
        return rslt_df.astype('int8')
    
   
```
